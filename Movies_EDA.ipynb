```python
#Finding all the title which are listed as Documentaries
documentaries_category = data.loc[data['listed_in'].str.contains("Documentaries", case=False)]
```

```python
documentaries_category.head()
```

```python
documentaries_category.tail()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = 'listed_in',
            y = 'release_year',
            #hue = 'listed_in',
            data = documentaries_category)
plt.xticks(rotation = 90)
plt.ylim(1940,2035)
plt.title("Variety of Documentary Released Per Year")
plt.show()
```

```python
#Finding all the work of Rajiv Chilaka
Rajiv_director = data[data['director'] == "Rajiv Chilaka"]
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = 'title',
            y = 'release_year',
            hue = 'listed_in',
            data = Rajiv_director)
plt.xticks(rotation = 90)
plt.ylim(2000,2025)
plt.title("Director Rajiv Chilaka's Work")
plt.show()
```

```python
dramas_inter_movies_category = data[data['listed_in'] == 'Dramas, International Movies']
```

```python
dramas_inter_movies_category.head()
```

```python
plt.figure(figsize=(23,9))
sns.barplot(x = 'country',
            y = 'release_year',
            #hue = 'listed_in',
            data = dramas_inter_movies_category)
plt.xticks(rotation = 90)
plt.ylim(1980,2025)
#set(plt.gca, 'XTick',{'India','United States','Italy','United Kingdom'})
plt.title("Drama Category released wordlwide")
plt.show()
```

```python
#To remove any NaN error exception
data.dropna(subset=["country"], inplace=True)

#Getting all the contents of Indian release
specific_country = data.loc[data['country'].str.contains("India", case=False)]
specific_country['release_year'].unique()
```

```python
#Fetching all the release from the year 2021
specific_year = specific_country.loc[(specific_country['release_year'] == 2021) |(specific_country['release_year'] == 2020) | (specific_country['release_year'] ==2019)]
specific_year['release_year'].unique()
```

```python
#Fetching a specific Category for movie
specific_category = specific_year.loc[specific_year['listed_in'].str.contains("Romantic", case=False)]
specific_category.info()
```

```python
plt.figure(figsize=(15,11))
sns.barplot(x = 'title',
            y = 'release_year',
            hue = 'rating',
            data = specific_category,
            palette="Blues_d")
plt.xticks(rotation = 90)
plt.ylim(2016,2022)
plt.title("Indian Romantic Movies ")
plt.show()
```
